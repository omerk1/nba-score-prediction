```python
# Quick exploration if database exists
import sqlite3

if data_path.exists():
    conn = sqlite3.connect(data_path)
    tables = pd.read_sql("SELECT name FROM sqlite_master WHERE type='table';", conn)
    print("Available tables:")
    for i, table in enumerate(tables['name'], 1):
        print(f"  {i}. {table}")
    conn.close()
```
